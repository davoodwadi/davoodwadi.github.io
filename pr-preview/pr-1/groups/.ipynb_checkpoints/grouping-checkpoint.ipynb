```python
import pandas as pd
```

```python
data = pd.read_csv("./MATH80629A Project Team Registration.csv")
```

```python
data
```

```python
dat = data[['Team name', 'List members name and their email address separated by comma']]
dat
```

```python
email_pattern = r'(\b[\w\.-]+@[\w\.-]+\b)'
dat['emails'] = data['List members name and their email address separated by comma']
dat['emails'].str.extractall(email_pattern).groupby(level=0).agg(', '.join)
```

```python
emails = dat['emails'].str.extractall(email_pattern).reset_index()
emails
```

```python
final = pd.merge(dat["Team name"],emails, right_on="level_0", left_index=True)
final
```

```python
# final[['Team name', '0']]
final.columns
```

```python
final[['Team name', 0]].to_csv("./groups.csv", index=False, header=False)
```
